```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv('covid_country_stats.csv')
```

```python
#Statistical Summary

#Top 5 coutries with most COVID-19 related deaths.
plt.figure(figsize = (12,10))
df_US = df.loc[df['Country/Region'].str.contains('US')]
df_US = df_US.groupby('Date')['Deaths'].sum()
df_US.plot(kind = 'line', label = 'US')

df_Ind = df.loc[df['Country/Region'].str.contains('India')]
df_Ind = df_Ind.groupby('Date')['Deaths'].sum()
df_Ind.plot(kind = 'line', label = 'India')

df_Brz = df.loc[df['Country/Region'].str.contains('Brazil')]
df_Brz = df_Brz.groupby('Date')['Deaths'].sum()
df_Brz.plot(kind = 'line', label = 'Brazil')

df_FNC = df.loc[df['Country/Region'].str.contains('France')]
df_FNC = df_FNC.groupby('Date')['Deaths'].sum()
df_FNC.plot(kind = 'line', label = 'France')

df_Tuk = df.loc[df['Country/Region'].str.contains('Turkey')]
df_Tuk = df_Tuk.groupby('Date')['Deaths'].sum()
df_Tuk.plot(kind = 'line', label = 'Turkey')


plt.title('Top 5 Countries With The Most COVID-19 Deaths',fontdict = {'fontsize': 15})
plt.xlabel('Date', fontdict = {'fontsize': 15})
plt.ylabel("Number Of Deaths", fontdict = {'fontsize': 15})
plt.legend()
```

```python
#Statistical Summary
df_pie = df.loc[(df['Recovered'] >= 80000)]
df_pie = df_pie.groupby('WHO Region')['Recovered'].mean()
df_pie.plot(kind = 'pie', radius = 1, figsize = (8,8))
plt.title('Regions with +80,000 Recoveries', fontdict = {'fontsize' : 20})
plt.tight_layout()
```

```python
#Pattern 

#During Summer cases started to spike 
plt.figure(figsize = (12,10))
df_US = df.loc[df['Country/Region'].str.contains('US')]
df_US = df_US.groupby('Date')['Deaths'].sum()
df_US.plot(kind = 'line', label = 'US')

df_Ind = df.loc[df['Country/Region'].str.contains('India')]
df_Ind = df_Ind.groupby('Date')['Deaths'].sum()
df_Ind.plot(kind = 'line', label = 'India')

df_Brz = df.loc[df['Country/Region'].str.contains('Brazil')]
df_Brz = df_Brz.groupby('Date')['Deaths'].sum()
df_Brz.plot(kind = 'line', label = 'Brazil')



plt.title('COVID-19 Cases Spike During Summer',fontdict = {'fontsize': 15})
plt.xlabel('Date', fontdict = {'fontsize': 15})
plt.ylabel("Number Of Cases", fontdict = {'fontsize': 15})
plt.legend()
```

```python
#Pattern 

#The Relationship Between Middle Eastern Countries and COVID-19
plt.figure(figsize = (12,10))
df_IRQ = df.loc[df['Country/Region'].str.contains('Iraq')]
df_IRQ = df_IRQ.groupby('Date')['Deaths'].sum()
df_IRQ.plot(kind = 'line', label = 'Iraq')

df_EGT = df.loc[df['Country/Region'].str.contains('Egypt')]
df_EGT = df_EGT.groupby('Date')['Deaths'].sum()
df_EGT.plot(kind = 'line', label = 'Egypt')

df_PKN = df.loc[df['Country/Region'].str.contains('Pakistan')]
df_PKN = df_PKN.groupby('Date')['Deaths'].sum()
df_PKN.plot(kind = 'line', label = 'Pakistan')



plt.title('COVID-19 & The Middle East',fontdict = {'fontsize': 15})
plt.xlabel('Date', fontdict = {'fontsize': 15})
plt.ylabel("Number Of Cases", fontdict = {'fontsize': 15})
plt.legend()
```
